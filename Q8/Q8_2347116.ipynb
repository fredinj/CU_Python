```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file=pd.read_csv("books_dataset.csv")
```

```python
missing_data=file.isnull()
```

```python
# display null values

null_counts = missing_data.sum()
for column, count in null_counts.items():
    print(f"Column '{column}': {count} null values")
```

```python
# replace save null values with 0

file['save'].fillna(0, inplace=True)
null_counts = file.isnull().sum()
for column, count in null_counts.items():
    print(f"Column '{column}': {count} null values")
```

```python
# use ffill on publisher

file.dropna(subset=['publisher'], inplace=True)
null_counts = file.isnull().sum()
for column, count in null_counts.items():
    print(f"Column '{column}': {count} null values")
```

```python
stars=file.groupby('stars')
stars['price'].mean()
```
